Support Vector Machine Algorithm for Regression (this could be done by using linear and polynomial kernel, we will start off from the first one)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVR
```

```python
df = pd.read_csv("ercot-dataset.csv", sep=',')
# to change the float to two decimal places following code is used
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print(df)
```

```python
# prints the high level information about the data
df.info()
```

```python
# gives additional statistical information
df.describe()
```

```python
# to check the total missing data on each features
# zero indicates there is no missing data values
df.isnull().sum()
df1 = df.drop(['Date', 'Month', 'Year'], axis=1)
df1 = df1.iloc[:400]
```

```python
# to see the correlation between features
# this helps to see how one parameters has effects on others
# lets see the plot of correlation data
sns.heatmap(df1.corr(), annot= True)
plt.rcParams["figure.figsize"]=(15,10)
plt.title('Correlation Matrix', fontsize=16)
plt.show()
```

Importing the required libraries

```python
# import sklearn libraries modules
from sklearn.model_selection import train_test_split
# this will select all the rows and columns except the last one from df and assign to x variable
# df1 does not contain the date in the first column in string format
x = df1.iloc[:,:-1].values
#x = df1.iloc[:,1].values
# quality is taken target variable and it in last column and is assigned to y variable
y = df1.iloc[:,-1].values
# encoding the categorical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [3])], remainder = 'passthrough')
x = ct.fit_transform(x)
```

Splitting the dataset into the training and test set

```python
# we will split the data into training and testing data using following code
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

training the support vector machine regression model

```python
# https://www.geeksforgeeks.org/support-vector-regression-svr-using-linear-and-non-linear-kernels-in-scikit-learn/
# create a SVR model with a linear kernel
svr = SVR(kernel='poly')
# trian the model on the data
svr.fit(x_train,y_train)
```

Predicting the test results

```python
y_pred = svr.predict(x_test)
```

visualization of the Random forest regression results graphically

```python
plt.plot(y_pred,y_test, '.')
#plt.plot(x_grid, regressor.predict(x_grid.reshape(-1,1)), color = 'blue')
plt.title('Support Vector Machine')
plt.xlabel('Actual Test Values')
plt.ylabel('Predicted System Load')
plt.show()
```

Plot the predicted values against the true values

```python
#plt.scatter(x_test, y_test, color = 'darkorange', label = 'data')
#plt.plot(y_test, y_pred, color = 'cornflowerblue', label = 'prediction')
plt.plot(y_test, color = 'blue', label = 'actual test data', marker='*', linestyle='--')
plt.plot(y_pred, color = 'red', label = 'predicted data', marker='o')
plt.legend()
plt.show()
```

Evaluating the model performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

the r2 score is better when there is less amount of training data. 
Removing the onehotencoder increased the r2 score thereby improving the model performance. If we have a lot of data to train the model, this algorithm does not suits well.
But, when we changed the kernel to polynomial, the r2 score drastically reduced to the -0.0130 from 0.9128 for sample data of 97 rows.

After several attempts to adjust the value of the sample, it was concluded that SVM with polynomial Kernel does not seem to work well for this kind of data.
